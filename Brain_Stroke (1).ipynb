```python
def outlier_detection_treatment(col):
    q1=df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    iqr= q3-q1
    lf= q1- 1.5*iqr
    uf= q3+ 1.5*iqr
    outliers= df[(df[col]<lf) | (df[col]>uf)]
    print("Percentage of outliers in", col ,outliers.shape[0]*100/df.shape[0])
    fig, ax= plt.subplots(1,2, figsize=(6,4))
    ax[0].boxplot(df[col])
    ax[1].hist(df[col])
    plt.suptitle(col)
    plt.show()
    df.loc[(df[col]<lf) | (df[col]>uf), col]= df[col].median()
```

```python
for col in num:
    print(outlier_detection_treatment(col))
    print("+++====+++")
```

# Encoding the data

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['ever_married']=le.fit_transform(df['ever_married'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cat
```

```python
df.drop(['smoking_status'],axis=1,inplace=True)
```

```python
df
```

```python
gender=OneHotEncoder(sparse_output=False)
worktype=OneHotEncoder(sparse_output=False)
Residencetype=OneHotEncoder(sparse_output=False)
```

```python
g_ohe=gender.fit_transform(df[['gender']])
work_ohe=worktype.fit_transform(df[['work_type']])
residency_ohe=Residencetype.fit_transform(df[['Residence_type']])
```

```python
df_gender= pd.DataFrame(g_ohe, columns= gender.categories_)
df_work=pd.DataFrame(work_ohe, columns=worktype.categories_)
df_residency=pd.DataFrame(residency_ohe, columns=Residencetype.categories_)
```

```python
df_gender.drop(['Female'],axis=1,inplace=True)
```

```python
df_work.drop(['children'],axis=1,inplace=True)
```

```python
df_residency.drop(['Rural'],axis=1,inplace=True)
```

```python
X= df.drop(['gender', 'work_type', 'Residence_type','stroke'], axis= 1)
y= df['stroke']
```

```python
X= pd.concat([X, df_gender, df_work, df_residency], axis=1)
```

```python
X.info()
```

```python
y
```

```python
x=X.values
y=y.values
```

# Spliting Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

# Applying Decision Tree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=5)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred_train=dt.predict(x_train)
y_pred_test= dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
```

```python
print('Accuracy for Training:',accuracy_score(y_train,y_pred_train))
print('Accuracy for Testing:',accuracy_score(y_test,y_pred_test))
```

```python
print('Confusion matrix for Training \n',confusion_matrix(y_train,y_pred_train),'\n')
print('Confusion matrix for Testing \n',confusion_matrix(y_test,y_pred_test))
```

# Decision Tree With Grid

```python
from sklearn.model_selection import GridSearchCV

grid={
    'max_depth':[10,20,30,40,50],
    "criterion":['gini','entropy'],
    'min_samples_leaf':[1,2,3,4,5,8,10],
    'min_samples_split':[2,5,10,15,20],
    "max_features":['sqrt','log2',None]
}
```

```python
gridsearch_dt=GridSearchCV(param_grid=grid,estimator=dt,cv=5,scoring='accuracy')
```

```python
gridsearch_dt.fit(x_train,y_train)
```

```python
y_grid_pred_train=gridsearch_dt.predict(x_train)
y_grid_pred_test=gridsearch_dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Accuracy for Training with grid:',accuracy_score(y_train,y_grid_pred_train))
print('Accuracy for Testing with grid:',accuracy_score(y_test,y_grid_pred_test))
```

```python
print('Confusion matrix for Training with grid \n',confusion_matrix(y_train,y_grid_pred_train),'\n')
print('Confusion matrix for Testing with grid \n',confusion_matrix(y_test,y_grid_pred_test))
```

# Applying Random Forest Classifier Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(random_state=5)
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
```

```python
grid_search = GridSearchCV(estimator=rfc, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
```

```python
grid_search.fit(x_train, y_train)
```

```python
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")
```

```python
best_rfc = grid_search.best_estimator_
```

```python
y_pred_train = best_rfc.predict(x_train)
y_pred_test = best_rfc.predict(x_test)
```

```python
print('Accuracy for RF Training:',accuracy_score(y_train,y_pred_train))
print('Accuracy for RF Testing:',accuracy_score(y_test,y_pred_test))
```
